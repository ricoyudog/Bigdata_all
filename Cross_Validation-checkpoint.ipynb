```python
############### 1 ###############
import pandas as pd

url = 'https://www4.comp.polyu.edu.hk/~xiaohuang/docs/CV_College.csv'
college = pd.read_csv(url, index_col=0)
print(college.head)
```

```python
############### 2 ###############
college.describe()
```

```python
############### 3 ###############
import seaborn as sns
sns.pairplot(college.iloc[:,:5])
```

```python
############### 4 ###############
sns.boxplot(x=college['Private'], y=college['Outstate'])
```

```python
############### 5 ###############
college = pd.get_dummies(college, drop_first=True)
print(college.head())
print(college.shape)
```

```python
############### 6 ###############
from sklearn.preprocessing import PolynomialFeatures
y = college['Apps']
X = college.iloc[:,1:]
X_ex = PolynomialFeatures(include_bias=False).fit_transform(X)
print(X_ex.shape)
```

```python
############### 7 ###############
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_ex, y, random_state=42)
print(X_train.shape[0])
```

```python
############### 8 ###############
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler = StandardScaler()
X_all = np.concatenate([X_train,X_test])
X_all = scaler.fit_transform(X_all)
X_train = X_all[0:X_train.shape[0], :]
X_test = X_all[X_train.shape[0]:, :]

print("Mean of all columns: {}".format(np.mean(X_all, axis=0)))
print("Standard deviation of all columns: {}".format(np.std(X_all, axis=0)))
print("Mean of all columns in training: {}".format(np.mean(X_train, axis=0)))
print("Standard deviation of all columns in training: {}".format(np.std(X_train, axis=0)))
```

```python
############### 9 ###############
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
regr = LinearRegression(fit_intercept=True).fit(X_train, y_train)

Apps_pred = regr.predict(X_test)
print("R^2 for Linear Regression: {}".format(r2_score(y_test, Apps_pred)))
```

```python
############### 10 ###############
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV, Ridge
alphas = 10**np.linspace(5,-2,100)*0.5
#print("alphas: {}".format(alphas))
regr = RidgeCV(alphas=alphas, store_cv_values=True).fit(X_train, y_train)

axmse = plt.gca()
axmse.plot(alphas, np.mean(regr.cv_values_, axis=0))
axmse.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('cross-validated MSE for training')
Apps_pred = regr.predict(X_test)
print("R^2 for ridge regression: {}".format(r2_score(y_test, Apps_pred)))
```

```python
############### 10 ###############
ridge = Ridge()
coefs = []
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(X_train, y_train)
    coefs.append(ridge.coef_)

ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('Standardized Coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.legend(X.columns[:4])
```

```python
############### 11 ###############
from sklearn.linear_model import LassoCV, Lasso
from sklearn.utils.testing import ignore_warnings
from sklearn.exceptions import ConvergenceWarning

@ignore_warnings(category=ConvergenceWarning)
def my_function():
  alphas = 10**np.linspace(5,-2,100)*0.5
  return LassoCV(alphas=alphas).fit(X_train, y_train)

regr = my_function()

axmse = plt.gca()
axmse.plot(alphas, np.mean(regr.mse_path_, axis=1))
axmse.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('cross-validated MSE for training')
Apps_pred = regr.predict(X_test)
print("R^2 for lasso regression: {}".format(r2_score(y_test, Apps_pred)))
print('Number of non-zero coefficients: {}. In total we have {} coefficients'.format(np.count_nonzero(regr.coef_), X_test.shape[1]))
```

```python
############### 11 ###############
coefs = []
@ignore_warnings(category=ConvergenceWarning)
def my_function():
  regr = Lasso().set_params(alpha=a)
  regr.fit(X_train, y_train)
  return regr.coef_

for a in alphas:
  coefs.append(my_function())

ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('Standardized Coefficients')
plt.title('lasso coefficients as a function of the regularization')
plt.legend(X.columns[:4])
```

```python
############### 12 ###############
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

forest_X = X_train[:,0:17]
scorelist = []
depths = list(range(1, 10)) 
for depth in depths:
  regr = RandomForestRegressor(max_depth=depth)
  scores = cross_val_score(regr, forest_X, y_train, scoring='neg_mean_squared_error', cv=10)
  scorelist.append(scores.mean() * -1)
print(scorelist)

axmse = plt.gca()
axmse.plot(depths, scorelist)
plt.xlabel('max depth')
plt.ylabel('10-fold cross-validated MSE for training')
print('Best max_depth selected by cv: {}'.format(depths[scorelist]))
regr = RandomForestRegressor(max_depth=(depths[scorelist])).fit(forest_X, y_train)
Apps_pred = regr.predict(X_test[:, 0:17])
print("R^2 for Random Forest regression: {}".format(r2_score(y_test, Apps_pred)))
```

```python
############### 13 ###############
from sklearn.linear_model import ElasticNetCV
from mpl_toolkits.mplot3d import Axes3D
from sklearn.utils.testing import ignore_warnings
from sklearn.exceptions import ConvergenceWarning
from matplotlib import cm

alphas = 10**np.linspace(3,-3,50)*0.5
@ignore_warnings(category=ConvergenceWarning)
def my_function():  
  return ElasticNetCV(alphas=alphas, l1_ratio=np.linspace(.05, 1, 20)).fit(X_train, y_train)
regr = my_function()
fig = plt.figure()
X, Y = np.meshgrid(alphas, np.linspace(.05, 1, 20))

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, regr.mse_path_.mean(axis=2), cmap=cm.coolwarm)
ax.set_xlabel('alpha')
ax.set_ylabel('l1_ratio')
ax.set_zlabel('mse');
Apps_pred = regr.predict(X_test)
print("R^2 for Elastic Net regression: {}".format(r2_score(y_test, Apps_pred)))
```
